```python
im2 = im[:,:,:3]
im2.shape 
```

```python
R = im2[:,:, 0]
G = im2[:, : , 1]
B = im2[: ,: , 2]
```

```python
plt.imshow(R , cmap = 'gray')
```

```python
plt.imshow(G , cmap = 'gray')
```

```python
fig , (ax1 , ax2 , ax3) = plt.subplots(1 , 3, figsize = (15,15))
ax1.imshow(R , cmap= 'gray')
ax2.imshow(G , cmap = 'gray')
ax3.imshow(B , cmap = 'gray')
```

```python
im3 = im2.copy()
im3[:,:,2]= 255
```

```python
plt.imshow(im3)
```

```python
im4 = im2.copy()
im4[200:400 , 200:400 , 0]= 10
im4[200:400 , 200:400 , 1] = 0
im4[200:400 , 200:400 , 2] = 200
plt.imshow(im4)
```

```python
imgray = 0.333*R + 0.333*G + 0.333*B 
plt.imshow(imgray , cmap= 'gray')
```

```python
#actual weightage while converting a rgb image to gray scale is not uniform because it does'nt seem to relistic 
imgray = 0.2989*R + 0.5870*G + 0.1140*B 
plt.imshow(imgray , cmap='gray')
```
